```python
import pandas as pd
```

```python
df = pd.read_csv('/content/customers.csv')
```

```python
df.describe()

rata_rata_umur = df['age'].mean()
print("Rata-rata umur:", rata_rata_umur)

modus_umur = df['age'].mode()
print("Modus umur:", modus_umur)

modus_negara = df['state'].mode()
print("Modus negara:", modus_negara)
```

```python
import matplotlib.pyplot as plt

# Tinggi
df['age'].hist()
plt.title("Distribusi umur customer (th)")
plt.show()

df['gender'].hist()
plt.title("Distribusi gender customer")
plt.show()

df['state'].hist()
plt.title("Distribusi kota customer")
plt.show()
```

```python
# Mean
group_mean_age_head = df.groupby("age").mean(numeric_only=True)
group_mean_age_head.sort_values(by='customer_id', ascending=False).head(5)

# Agg lengkap: Menghitung jumlah customer per kelompok umur, mengurutkan berdasarkan jumlah customer terbanyak, dan menampilkan 5 teratas.
group_agg_age = df.groupby("age").agg({
    'customer_id': ['count']
})
group_agg_age.sort_values(('customer_id', 'count'), ascending=False).head(5)
```

```python
# Korelasi
corr_age_customer_id = df[['age', 'customer_id']].corr()
print(corr_age_customer_id)

# Kovarian
cov_age_customer_id = df[['age', 'customer_id']].cov()
print(cov_age_customer_id)
```
